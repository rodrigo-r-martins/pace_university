```python
# Program used to replicate experiments on databases that were extracted from Twitter
import os

input_filename = os.path.join(os.path.expanduser('~'), 'OneDrive', 'Desktop', 'Pace', 'CS619', 'Chapter06', 'Data', 'python_tweets.json')
labels_filename = os.path.join(os.path.expanduser('~'), 'OneDrive', 'Desktop', 'Pace', 'CS619', 'Chapter06', 'Data', 'python_classes.json')
replicable_dataset = os.path.join(os.path.expanduser('~'), 'OneDrive', 'Desktop', 'Pace', 'CS619', 'Chapter06', 'Data', 'replicable_dataset.json')
```

```python
# Loading the tweets and labels
import json

tweets = []
with open(input_filename) as inf:
    for line in inf:
        if len(line.strip()) == 0:
            continue
        tweets.append(json.loads(line))
        
if os.path.exists(labels_filename):
    with open(labels_filename) as inf:
        labels = json.load(inf)
```

```python
# Creating a dataset
dataset = [(tweet['id'], label) for label, tweet in zip(labels, tweets)]
print(dataset)
```

```python
# Saving the results in our file
with open(replicable_dataset, 'w') as outf:
    json.dump(dataset, outf)
```
